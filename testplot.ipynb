```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4,8,12,16,20])
y1_mean = np.array([4682.19,	4789.72875,	4847.804167,	4917.538125,	5016.2175])
y1_std = np.array([0.636327,63.415373,68.622458,205.830309,390.84450])

#y2_mean = np.array([])

# create the ribbon plot
plt.plot(x, y1_mean, label='y1 mean')
#plt.plot(x, y2_mean, label='y2 mean')
plt.fill_between(x, y1_mean-y1_std, y1_mean+y1_std, alpha=0.2, label='y1 std')
#plt.fill_between(x, y2_mean-y2_std, y2_mean+y2_std, alpha=0.2, label='y2 std')
#plt.ylim(0,6000)
plt.legend()
plt.show()
```
