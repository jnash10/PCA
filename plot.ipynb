```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=1
b=0.5
def f1(x,y):
    return a*(1-np.e**(-b*((x)**2 + (y)**2)))

vf1 = np.vectorize(f1)
```

```python
X = np.arange(-5, 5, 0.01)
Y = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
```

```python
vf1(X,Y)
```

```python
from matplotlib import cm
```

```python
from matplotlib.ticker import LinearLocator
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(X, Y, vf1(X,Y), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.xlabel('x', fontsize=24)
plt.ylabel('y',  fontsize=24)
ax.set_zlabel('|F|', rotation=0,  fontsize=24)
plt.savefig('/media/storage/Agam/pca/plots/fields/field_attractive.eps', format='eps')

plt.show()
```

```python
a=1
b=0.5
def f(x,y):
    return a*(1*np.e**(-b*((x)**2 + (y)**2)))

vf2 = np.vectorize(f)
```

```python
X = np.arange(-5, 5, 0.01)
Y = np.arange(-5, 5, 0.01)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
```

```python
vf2(X,Y)
```

```python
from matplotlib import cm
```

```python
from matplotlib.ticker import LinearLocator
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(X, Y, vf2(X,Y), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
plt.xlabel('x', fontsize=24)
plt.ylabel('y',  fontsize=24)
ax.set_zlabel('|F|', rotation=0,  fontsize=24)
plt.savefig('/media/storage/Agam/pca/plots/fields/field_repulsive.eps', format='eps', bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))
surf = ax.plot_surface(X, Y, vf1(X-2,Y-2)+vf2(X+2,Y+2), cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

plt.savefig('/media/storage/Agam/pca/plots/fields/both.png')
plt.show()
```
